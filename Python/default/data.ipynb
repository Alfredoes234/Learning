```python
df = pd.read_csv('vgsales.csv')
df
```

```python
df.info()
df = df.drop(['Rank'], axis=1)
new_names = {'Global_Sales':'Global'}
df.rename(columns=new_names, inplace=True)
df.head()
```

```python
Nintendo = df[df.Publisher == 'Nintendo']
Nintendo
```

```python
Activision = df[df.Publisher == 'Activision']
Activision
```

```python
plt.bar(Nintendo.Year,Nintendo.Global)
plt.bar(Activision.Year,Activision.Global)
plt.xlabel('Game release year')
plt.ylabel('Global sales')
plt.title('Nintendo Vs Activision Global sales')
plt.legend(['Nintendo','Activision'])
plt.show()
```

```python
ef = pd.read_csv('2019.csv')
ef.info()
ef = ef.drop(['Overall rank'], axis=1)
new_names = {'Country or region':'Country', 'Score':'Happiness', 'GDP per capita':'GDP_per_capita', 'Social support':'Social', 'Healthy life expectancy':'Health'}
ef.rename(columns=new_names, inplace=True)
ef.head()
```

```python
plt.scatter(ef.Happiness,ef.GDP_per_capita)
plt.xlabel('Happiness')
plt.ylabel('GDP Per Capita')
plt.title('Correlation between happiness and the GDP')
plt.show()
```

```python
plt.hist(ef.Generosity,bins=16,linewidth=1,edgecolor='k')
plt.title('Generosity distribution')
plt.xlabel('gen')
plt.show()
```

```python
Uk = ef[ef.Country == 'United Kingdom']
Nus = ef[ef.Country == 'United States']
Russia = ef[ef.Country == 'Russia']

all_dat = pd.concat([Uk, Nus , Russia])

country_data = all_dat['Country']
Money_avg = all_dat['GDP_per_capita']
plt.pie(Money_avg, labels=country_data,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('GDP per capita')
plt.show()
```
